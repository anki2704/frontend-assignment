**Q1. Create a simple page with some div tags and show different ways to add CSS as well as what happens
when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the
project where required**


```

**q1.html**

<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Q1 - CSS Priority Example</title>

    <!--  Internal CSS -->
    <style>
        /* Internal CSS for div styles */
        .box {
            width: 200px;
            height: 100px;
            background-color: lightblue;
            border: 2px solid blue;
            margin: 10px;
            color: darkblue;
        }
    </style>

    <!--  Linking External CSS -->
    <link rel="stylesheet" href="q1.css">
</head>

<body>

    <!--  Inline CSS applied (highest priority) -->
    <div class="box" style="background-color: yellow; color: black;">
        Inline CSS Example
    </div>

    <!--  Will use Internal CSS unless overridden by External CSS -->
    <div class="box">
        Internal & External CSS Example
    </div>

    <!-- Only External CSS will apply here -->
    <div class="container">
        External CSS Only
    </div>

    <!--  Comment Example -->
    <!-- This div is used to show how inline, internal, and external styles override each other -->
</body>

</html>


**q1.css**

/* External CSS File */

/* Styling the second div (class: box) */
.box {
    background-color: lightgreen; /* Will be overridden by inline CSS (for first div) */
    color: darkgreen;
    border: 2px dashed green;
}

/* Styling the third div separately */
.container {
    width: 200px;
    height: 100px;
    background-color: lightcoral;
    border: 2px solid darkred;
    margin: 10px;
    color: white;
}








**Q2. Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using
class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and
explain how you've named the classes**

```
**q2.html**

<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Q2 - Paragraph Styling with BEM</title>
    <link rel="stylesheet" href="q2.css">
</head>

<body>

    <!--  BEM Naming Convention:
         Block: paragraph
         Element: (not needed here)
         Modifier: --variant1, --variant2, etc.
    -->

    <p class="paragraph paragraph--intro">
        This is the introduction paragraph, styled using BEM.
    </p>

    <p class="paragraph paragraph--highlight">
        This is the highlighted paragraph with a different background.
    </p>

    <p class="paragraph paragraph--note">
        This is a note paragraph with italic text styling.
    </p>

    <p class="paragraph paragraph--alert">
        This paragraph is styled to look like an alert or warning.
    </p>

</body>

</html>



**q2.css**
/* Base Block Style (BEM Block) */
.paragraph {
    font-size: 16px;
    margin: 12px 0;
    padding: 10px;
}

/*  Modifier 1: Introduction */
.paragraph--intro {
    color: darkblue;
    font-weight: 600;
    border-left: 4px solid darkblue;
    background-color: #e6f0ff;
}

/*  Modifier 2: Highlight */
.paragraph--highlight {
    background-color: #fff3b0;
    color: #6b4f00;
    border-left: 4px solid #e0b800;
}

/*  Modifier 3: Note */
.paragraph--note {
    font-style: italic;
    color: #444;
    background-color: #f5f5f5;
    border-left: 4px solid gray;
}

/*  Modifier 4: Alert */
.paragraph--alert {
    background-color: #ffd7d7;
    color: #b30000;
    border-left: 4px solid red;
    font-weight: bold;
}
