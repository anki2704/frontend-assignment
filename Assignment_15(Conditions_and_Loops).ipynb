**Q1. What are conditional statements? Explain conditional statements with syntax and examples.**

Ans.Conditional statements in JavaScript are used to make decisions in a program. They allow the execution of different blocks of code based on whether a given condition evaluates to true or false.

Types of Conditional Statements
1. **if Statement**
: Executes a block of code when the specified condition is true.
  
 **Syntax:**
  
   if ( condition){ // statement; }

  **Example:**
  
   let num = 7;
  
  if ( x>5 ){ console.log( number is greater than 5 ); }


2. **if–else Statement:**
Executes one block if the condition is true, otherwise executes another block.

**Syntax:**

 if (condition) {
    // if statement;
}

else {
    // else statemnet;
}

 **Example:**

  let num = 7;

if (num % 2 === 0)

 {
    console.log("Even");
}

else {
    console.log("Odd");
}


3. **if-else–if  ladder statement:**

Used to check multiple conditions.

**Syntax:**

if (condition1) {
    // code
}

else if (condition2) {
    // code
}

else {
    // default code
}


**Example:**

let marks = 80;

if (marks >= 75)

{
    console.log("Distinction");
}

 else if (marks >= 50) {
    console.log("Pass");
}

 else {
    console.log("Fail");
}




4. **Nested if statement:**
We can nest if else statements inside another if else statements.

**Syntax:**

if (condition1)

{
    if (condition2)
    {
        // code to be executed when both conditions are true
    }
    else {
        // code when condition2 is false
    }
}

else {
    // code when condition1 is false
}


**Example:**  

let age =17;
let hasId = true ;
 if (age>= 18){
    if (hasId){
      console.log ( " Entry allowed");
    } else{
      console.log("Id required");

 }

**Q3. What are loops, and what do we need them? Explain different types of loops with their syntax and
examples.**

**Ans.** Loops are control structures used to execute a block of code repeatedly as long as a specified condition remains true. They help in reducing repetition of code and make programs more efficient and easier to maintain.

**Need for Loops:**

Loops are required to perform repetitive tasks, process arrays and data collections, and execute statements multiple times without writing the same code again.

**Types of Loops:**

**1. for Loop:**

Used when the number of iterations is known.

**Syntax:**

for (initialization; condition; update) {
    // statements
}


*Example:*

for (let i = 1; i <= 5; i++) {
    console.log(i);
}

**2. while Loop:**

Executes code as long as the condition is true.

**Syntax:**

while (condition) {
    // statements
}


*Example:*

let i = 1;
while (i <= 5) {
    console.log(i);
    i++;
}

**3. do–while Loop:**

Executes the loop body at least once.

**Syntax:**

do {
    // statements
} while (condition);


*Example:*

let i = 1;
do {
    console.log(i);
    i++;
} while (i <= 5);
