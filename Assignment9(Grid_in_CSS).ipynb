**Q3.Explain the use of grid-auto-row and grid-auto-column using code examples.**

**Ans:**


**1.grid-auto-rows:**
 (initial value - auto)
 grid-auto-rows property specifies the height of rows that are automatically created when there is no explicit
row definition.

 If there are any explicitly defined rows in the grid using the "grid-template-rows" property, then those
rows will have their own heights, and the "grid-auto-rows" property will only apply to any additional rows that
are created automatically.


**2.grid-auto-columns:**
 (initial value - auto)
 Similarly, the grid-auto-columns property specifies the width of columns that are automatically created when
there is no explicit column definition.

```


<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Q4</title>
    <style>
        .container{
            display: grid;
            grid-template-columns: repeat(6,120px);
            grid-template-rows: repeat(2,75px);
            gap: 1em;
            
        }
        .box{
            border: 1px solid black;
            border-radius: 5px;
            padding: 10px;
            font-weight: 500;
            font-size: 30px;
            align-content: center;
        }
        .box1{
            background-color: rgb(48, 45, 45);
            color: white;
            grid-row: 2/3;

        }
        .box2{
            background-color: rgb(187, 185, 185);
            grid-row: 2/3;

        }
        .box3{
            background-color: rgb(48, 45, 45);
            color: white;
    
        }
        .box4{
            background-color: rgb(187, 185, 185);
        }
        .box5{
            background-color: rgb(48, 45, 45);
            color: white;
    
        }
        .box6{
            background-color: rgb(187, 185, 185);
        }
        .box7{
            background-color: rgb(48, 45, 45);
            color: white;
        }
        .box8{
            background-color: rgb(187, 185, 185);
        }

    </style>

</head>
<body>
    <div class="container">

        <div class="box box1">1</div>
        <div class="box box2">2</div>
        <div class="box box3">3</div>
        <div class="box box4">4</div>
        <div class="box box5">5</div>
        <div class="box box6">6</div>
        <div class="box box7">7</div>
        <div class="box box8">8</div>

</div>
       
    </div>
</body>
</html>

**Q5.Explain the difference between justify-items and justify-self using code examples.**

**Ans**

**1.justify-items:**
 (default value - stretch)
 This property allows us to set the default way of align each item in their respective grid cells horizontally.
 Now, we will see how items get arranged using each possible value of justify-items.

 Values

start  Left

end  Right

center  Center

stretch (default)  Fill available width


**2.justify-self:**Aligns a SINGLE grid item horizontally.

Values->
start; end ; center ; stretch (default)
