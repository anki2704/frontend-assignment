**Q1.  Describe the main differences between the CSS Flexbox layout model and the CSS Grid layout model. When
would you choose to use one over the other?**

**Ans.**

**Flexbox:**


1.Layout Dimension:Flexbox is one-dimensional and works in either a row or a column.

2.Content Flow vs Structured Layout:Flexbox is content-driven. The items flow based on the content size.

3.Positioning Control:
Flexbox gives control along a single axis using properties like justify-content and align-items.

4.Use Cases:
Flexbox is ideal for small components like navbars, buttons, or aligning items in a single row/column.


5.Choosing Between Them:Use Flexbox when the layout is in one direction.






**CSS Grid:**

1.Layout Dimension:
Grid is two-dimensional and handles both rows and columns at the same time

2.Content Flow vs Structured Layout:
Grid is layout-driven. You define a structure first, and then place items into it.

3.Positioning Control:Grid allows precise placement using grid-template-rows, grid-template-columns, and grid-area.

4.Use Cases:Grid is better for page or section layouts like headers, sidebars, galleries, or dashboards.

5.Choosing Between Them:Use Grid when the layout needs both rows and columns.

**Q2.Explain the role of the following key properties in the Flexbox layout model**


**1.justify-content**

**2.align-item**

**3.gap**

**4.flex-direction**

**5.flex-wrap**

**Ans.**

1.justify-content :- The justify-content property is used to align and distribute flex items along the main axis inside a flex container.

2.align-item:- The align-item property is used to align and distribute flex items along the cross axis inside a flex container.

3.gap:-  the gap property srts the space between the  felx items without using margins. It helps maintain equal spacing between elements.

4.flex-direction:- This property defines the direction of the main axis and determines how items are placed.

Values:

-row (default â€“ left to right)

-column (top to bottom)

-row-reverse

-column-reverse

5.Flex wrape:- the wrape property  controls whether items should stay in a single line or wrap to the next line when space is insufficient.
