**Q1.Create a simple webpage with a header, a main content section, and a footer. Apply a media query that
changes the background color of the header to blue when the screen width is less than 600px.**

Screenshot 2025-11-25 192246.png

```


<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Media Query</title>
    <style>
        *{
            padding: 0;
            margin: 2px 0;
        }
        header
        {
            width: 100%;
            height: 100px;
            background-color: rgb(206, 206, 206);
            display: flex;
            justify-content: center;
            align-items: center;
        }
        main{
            margin: 10px;
            height: 200px;
        }
        footer{
            margin: 10px;
           
        }
        @media screen and (max-width:600px) {
            header{
                background-color: blue;
                
            }
        }
    </style>
</head>
<body>
    <header>
        <h1>Resposive Header</h1>
    </header>
    <main>
        <H5>Main Content</H5>
    </main>
    <Footer>
        <h4>Footer</h4>
    </Footer>
</body>
</html>

**Q2.Create an image gallery with three images in a row. Use media queries to adjust the layout to two images in a
row for screens smaller than 800px and one image in a row for screens smaller than 500px.For image use this link (Image Link)**

Screenshot 2025-11-25 192504.pngScreenshot 2025-11-25 192456.png

```

**Q3.Create a navigation bar with five links. Apply a media query that converts the navigation into a vertical list for
screens smaller than 768px and also changes the background color to gray.**

Screenshot 2025-11-25 192556.png

```

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>responsive navbar</title>
    <style>
        body{
            margin: 10px 10px;
        }
        nav {
         
            background-color: black;
        }
        nav li{
            width: 100%;
            height: 50px;
            display: flex;
            justify-content: center;
            align-content:space-around;
            align-items: center;
        }
       
        ul{
            display: flex;
            list-style: none;
            padding: 0px;
           
        
        }
        ul li a{
            color: white;
            text-decoration: none;
        }
        @media screen and (max-width:768px){
            ul{
                display: flex;
                flex-direction: column;
            }
            nav{
                background-color: gray;
            }
        }
       
    </style>
</head>
<body>
    <nav>
        <ul>
            <li><a href="home.html">Home</a></li>
            <li><a href="about.html">About</a></li>
            <li><a href="service.html">Services</a></li>
            <li><a href="portfolio.html">Portfolio</a></li>
            <li><a href="contact.html">Contact</a></li>
        </ul>
    </nav>
    
</body>
</html>

**Q4.Create a grid based layout containing four grid items which arranges them as the screen size changes.**Screenshot 2025-11-25 192655.png

```
